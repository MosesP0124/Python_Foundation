```python
#file을 다룰때는 항상 cursor의 위치를 신경써야합니다.
with open("test.txt", "w+") as f:
    f.write("test")
    f.seek(0)
    db = f.read()
    print(db)

with open("test.txt", "r+") as f: # 대상 file이 없으면 error
    f.write("b") # cursor 위치가 맨 앞부분에 위치
    f.seek(0)
    db = f.read()
    print(db)

with open("test.txt", "a+") as f: # 대상 file이 없으면 생성
    f.write("b") # cursor 위치가 파일의 맨 마지막 부분에 위치
    f.seek(0)
    db = f.read()
    print(db)
```

```python
# 제네레이터 defualt 설정 가능
def test(num):
    for x in range(num):
        yield x

tt = test(2)
print(next(tt))
print(next(tt, "finished")) # return 1
print(next(tt, "finished")) # return "finished"
```

```python
# lambda tuple 생성해서 사용하기
def test(x):
    return lambda y: x+y, lambda y: x+y+1, lambda y: x+y+2

a = test(5)

print(a[1](1))
```
